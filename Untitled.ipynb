```python
import pandas as pd

data = pd.read_csv('static/doc/survey.csv', encoding='utf8', index_col='이름')
data
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(font="Malgun Gothic", 
        rc={"axes.unicode_minus":False},
        style='darkgrid')
```

```python
nrow=7
ncol=2

plt.figure(figsize=(20, 30))
col_name = data.columns
for i in range(1, 15):
    plt.subplot(nrow,ncol,i)
    sns.countplot(data=data, x=data[col_name[i]])
plt.show()
```
